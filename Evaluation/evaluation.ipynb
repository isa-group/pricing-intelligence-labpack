```python
# Ejemplo: generar evaluation report usando el script creado
from pathlib import Path
import subprocess, sys

input_path = Path('data') / 'experiment_results_gpt_5_mini.json'
out_dir = Path('data')
outfile = 'evaluation_report_gpt_5_mini.json'
cmd = [sys.executable, 'generate_evaluation_report.py', '--input', str(input_path), '--output_dir', str(out_dir), '--outfile', outfile]
print('Ejecutando:', ' '.join(cmd))
subprocess.run(cmd, check=True)
```
